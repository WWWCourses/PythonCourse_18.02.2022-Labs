```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#inline magic to show figure inline:
%matplotlib inline
```

```python
# DataSet1:
# x1 = np.random.randint(1,11, 10)
# y1 = np.random.randint(20,31, 10)
x1 = np.arange(1,11)
y1 = np.arange(11,21)

# y1.shape
```

```python
# DataSet2:
x2 = np.arange(-5,6)
y2 = np.arange(-5,6)
```

```python
# plot line using State-based interface:
plt.plot(x1,y1,label='DataSet1')
plt.plot(x2,y2,label='DataSet2')
plt.legend()
# customize Figure attributes:
plt.title('My first line')
plt.xlabel('X',labelpad=50,loc="right")
plt.ylabel('Y',labelpad=20,loc="top")
_=plt.xticks(np.array(np.arange(-5,11)))
# _=plt.xticks(np.array([-5,5,10]),['a','b','c'])



```

```python
# plot line using OOP interface:
fig, [ax1,ax2] = plt.subplots(2)

ax1.plot(x1,y1,'or') #'[marker][line][color]'
ax1.plot(x2,y2,'g^')
ax1.set_title('My first line')
ax1.set_xlabel('X',labelpad=50,loc="right")
ax1.set_ylabel('Y',labelpad=20,loc="top")
```

```python
fig, [ax1,ax2] = plt.subplots(1,2)
ax1.scatter(x1,y1)
```

```python
# demo for pie plot:
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0.4, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels)

plt.show()
```
